```python
# Label Encoding the data
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

```python
newdf=dftrain.apply(le.fit_transform)
newtest=dftest.apply(le.fit_transform)
```

```python
newdf.shape,newtest.shape
```

```python
x_train=newdf.drop(['y'],axis=1)
```

```python
x_train.shape,newtest.shape
```

```python
# Performing Principal Componenet Analysis on the label encoded data for dimentionality reduction
from sklearn.decomposition import PCA
```

```python
sklearn_pca=PCA(n_components=10)
output_pca=sklearn_pca.fit_transform(x_train)
output_pca
```

```python
test_transformed = sklearn_pca.transform(newtest)
```

```python
explained_variance=sklearn_pca.explained_variance_ratio_
explained_variance
```

```python
# performing xgboost on label encoded and dimentionally reduced data
import xgboost as xgb
model=xgb.XGBClassifier(random_state=1, learning_rate=0.01)
model.fit(output_pca, newdf['y'])
```

```python
# making prediction on the test set.
preds=model.predict(test_transformed)
```
